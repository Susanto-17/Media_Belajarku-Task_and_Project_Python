```python
items = []
while True:
    print("------------------ Selamat Datang di Supermarket Media Belajarku ------------------")
    print("1. Tambah items \n2. Lihat items \n3. Beli items \n4. Cari items \n5. Edit items \n6. Keluar")
    pil = int(input("Input pilihan angka : "))

    if pil == 1 :
        print('------------------Tambah items------------------')
        item = {}
        item['nama'] = input('Nama Item : ')
        while True:
            try:
                item['jumlah'] = int(input('Jumlah item : '))
                break
            except ValueError:
                print('Input jumlah harus angka')
        while True:
            try:
                item["harga"] = int(input("Harga Rp : "))
                break
            except ValueError:
                print('Input Harga harus angka')
        print('Item berhasil ditambah.')
        items.append(item)
            

    elif pil == 2 :
        print('------------------Lihat Items------------------')
        print('Jumlah items dalam inventori : %d ' %len(items))
        if len(items) != 0:
            print('Item yang tersedia dalam supermarket.')
            for item in items:
                for key, value in item.items():
                    print("%s : %s " %(key, value))

    elif pil == 3 :
        print('------------------Beli items------------------')
        print(items)
        purchase_item = input('Pilih item yang ingin dibeli? Input nama item : ')
        purchase_quantity= int(input('Input jumlah item : '))
        for item in items:
            if purchase_item.lower() == item['nama'].lower() :
                if item['jumlah'] != 0 :
                    if purchase_quantity <= item['jumlah']:
                        print('Harga Barang Rp. %d di Counter.' %(item['harga']* purchase_quantity))
                        item['jumlah'] -= purchase_quantity
                    else:
                        print("Jumlah item tidak cukup")

                else: 
                    print('tidak ada stock item.')

    elif pil == 4 :
        print('------------------Cari items------------------')
        find_item = input('Input nama items yang dicari di inventory : ')
        for item in items:
            if item['nama'].lower() == find_item.lower():
                print('Nama item ' + find_item + ' berikut detail itemnya')
                print(item)
            else:
                print('item tidak ditemukan.')

    elif pil == 5 :
        print('------------------edit items------------------')
        item_name = input('Input nama item yang ingin diedit : ')
        for item in items:
            if item_name.lower() == item['nama'].lower():
                print('Ini keterangan item ' + item_name)
                print(item)
                item['nama'] = input('Nama item : ')
                while True:
                    try:
                        item['jumlah'] = int(input('Jumlah item : '))
                        break
                    except ValueError:
                        print('Jumlah harus angka')
                while True:
                    try:
                        item['harga'] = int(input('Harga Rp : '))
                        break
                    except ValueError:
                        print('Harga harus angka')
                print('Item berhasil diupdate.')
                print(item)
            else:
                print('Item tidak ditemukan')
                
    elif pil == 6 :
        print('------------------ keluar ------------------')
        break

    else: 
         print('Input tidak sesuai pilihan')
```
