# Pemesanan Tiket Pesawat

```python
class Garuda:
    def __init__(self):
        self.asal = ["Jakarta", "Surabaya", "Medan", "Bandung", "Semarang"]
        self.tujuan = ["Medan", "Jakarta", "Surabaya", "Jakarta", "Surabaya"]
        self.harga = ["2.000.000", "1.100.000", "2.800.000", "850.000"]
        self.tmp_jumlah = []
        self.tmp_harga = []
    
    def tmbh_ga_destinasi(self):
        new_asal = input("Input dari :")
        new_tujuan = input("Input tujuan :")
        new_harga = input("Input Harga")
        self.asal.append(new_asal)
        self.tujuan.append(new_tujuan)
        self.harga.append(new_harga)
        pass
    
    
    def harga_ga_tiket(self, kode, jumlah):
        ttl_harga = 0
        self.tmp_jumlah.append(jumlah)
        print("="*45)
        
        tmp_harga = self.harga[int(kode)-1]
        ttl_harga = int(tmp_harga.replace('.', '')) * int(jumlah)
        self.tmp_harga.append(ttl_harga)
        
        return ttl_harga
    
    def total_harga_ga_tiket(self):
        total_harga = 0
        for i in range(len(self.tmp_harga)):
            total_harga += self.tmp_harga[i]
        print("total harga : ", total_harga)
        return total_harga
    
    def reset_ga_pesanan(self):
        self.tmp_asal.clear()
        self.tmp_tujuan.clear()
        self.tmp_harga.clear()
        pass
    
    def cetak_jadwal_garuda(self):
        for i, (asal, tujuan, harga) in enumerate(zip(self.asal, self.tujuan, self.harga)):
            print("Kode Penerbangan : ", i+1, asal, "==>", tujuan, "==>", harga)
        pass
```
