```python
class menu_makan:
    def __init__(self):
        self.list_nama_makanan = ["1. Paket Ayam           Rp 10.000",
                                 "2. Paket Telor Penyet    Rp 10.000",
                                 "3. Paket Nasi Goreng     Rp 10.000",
                                 "4. Bakmi Goreng          Rp 10.000",
                                 "5. Bakmi kuah            Rp 10.000"]
        
        self.list_nama_minuman = ["1. Es teh               Rp 3.000",
                                 "2. Es Jeruk              Rp 5.000",
                                 "3. Kopi Susu             Rp 5.000",
                                 "4. Jus Mangga            Rp 8.000",
                                 "5. Jus Alpukat           Rp 8.000"]
        
        self.list_harga_makanan = [10000, 10000, 10000, 10000, 10000]
        self.list_harga_minuman = [3000, 5000, 5000, 8000, 8000]
        
        self.indeks_makanan = []
        self.indeks_minuman = []
        
        self.tmp_makanan = []
        self.tmp_minuman = []
        
        self.harga = []
        
    def print_list_makanan(self):
        for nama_makanan in self.list_nama_makanan:
            print(nama_makanan)
        print("="*40)
        pass
    
    def print_list_minuman(self):
        for nama_minuman in self.list_nama_minuman:
            print(nama_minuman)
        print("-"*40)
        pass
        
    def tambah_menu_makanan(self, makanan):
        self.list_nama_makanan.append(makanan)
        pass
    
    def tambah_menu_minuman(self, minuman):
        self.list_nama_minuman.append(minuman)
        pass
    
    def print_menu_yg_dipesan(self):
        print("")
        print("Makanan yang Anda pesan : ")
        for makan in self.indeks_makanan:
            print(self.list_nama_makanan[makan])
        print("")    
        print("Minuman yang Anda pesan : ")
        for minum in self.indeks_minuman:
            print(self.list_nama_minuman[minum])
        
        pass
        
    def tmp_pesan_makan(self, nama_makanan):
        self.indeks_makanan.append(nama_makanan)
        pass
    
    def tmp_pesan_minum(self, nama_minuman):
        self.indeks_minuman.append(nama_minuman)
        pass
    
    def total_harga(self):
        for mkn in self.indeks_makanan:
            self.harga.append(self.list_harga_makanan[mkn])
        for mnm in self.indeks_minuman:
            self.harga.append(self.list_harga_minuman[mnm])
            
        return self.harga
```

```python
menu = menu_makan()

nama_customer = input("Masukkan Nama Pemesan : ")
print("Halo ", nama_customer, " Silahkan pilih menu :")
print()
print("Menu Makanan")
menu.print_list_makanan()
    
print("Menu Minuman")
menu.print_list_minuman()
    
sign = True    
while True:
    kode_menu = int(input("Silahkan pilih menu : \n1. Makanan \n2. Minuman \n3. keluar: "))
        
    if kode_menu == 1:
        print("---------------Menu Makanan------------------")
        menu.print_list_makanan()
            
        while True:
            try:
                kode_makanan = int(input("Silahkan pilih no makanan anda atau exit(99): "))
                if kode_makanan == 99:
                    print("Keluar program")
                    break
                else:
                    menu.tmp_pesan_makan(kode_makanan-1)
            except ValueError:
                print("Kode yang diinput salah")
                    
                
    elif kode_menu == 2:
        print("---------------Menu Minuman------------------")
        menu.print_list_minuman()
            
        while True:
            try:
                kode_minuman = int(input("Silahkan pilih no minuman anda atau exit(99): "))
                if kode_minuman == 99:
                    break
                else:
                    menu.tmp_pesan_minum(kode_minuman-1)
            except VelueError:
                 print("Kode yang diinput salah")
                    
    elif kode_menu == 3:
        break
            
    else:
        print("Menu yang anda pilih salah, Silahkan pilih menu kembali")
    
menu.print_menu_yg_dipesan()

print("-"*40)
ttl_harga = menu.total_harga()
print("Total Harga :        ", sum(ttl_harga))
```
